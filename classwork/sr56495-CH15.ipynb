```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
x_start, x_end = 0, 2

first_run = None 

for h in np.arange(0.01, 1.01, 0.05):

    num_steps = int((x_end - x_start) / h)

    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)

    # Initial condition
    y_values[0] = 1  # y(0) = 1

    # Function representing dy/dx = -2y
    def f(x, y):
        return -2 * y

    # Implement Euler's method
    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])

    y_exact = np.exp(-2 * x_values)

    error = np.abs(y_values - y_exact)
    total_error = np.sum(error)

    print(f"h = {h:.2f}, Total Error = {total_error:.5f}")

    if first_run is None:
        first_run = y_values.copy()
    else:
        min_len = min(len(first_run), len(y_values))
        compare_error = np.abs(y_values[:min_len] - first_run[:min_len])
        print(f"   Error vs first run = {np.sum(compare_error):.5f}")

    # Plot each run
    plt.plot(x_values, y_values, label=f"h={h:.2f}")

x_fine = np.linspace(x_start, x_end, 100)
plt.plot(x_fine, np.exp(-2 * x_fine), 'k', linewidth=3, label="Exact")

plt.xlabel('x')
plt.ylabel('y')
plt.title("Approximation of dy/dx = -2y using Euler's Method")
plt.legend()
plt.show()
```

We can observe given the results how as step size increases, so does error. Therefore, smaller step sizes give more accurate solutions with Euler's method. This is because the approximation of the derivative with finite steps gets less accurate as larger steps are used
