```python
import numpy as np

piEstimate = 0.0

#leibniz series setup
for n in range(5000):
    term = ((-1)**n) / (2*n + 1)
    piEstimate += term
    
    if (n+1) % 100 == 0:
        approximatePi = 4*piEstimate
        
        #calculation for relative error
        relativeError = abs(np.pi - approximatePi) / np.pi
        print(f"at terms: {n+1}, relative error:{relativeError}")
        #as each term is added, the error gets slightly smaller
```
