```python
import numpy as np

def diagonally_dominant_check(A):
    n = len(A)
    
    for i in range(n):
        diagonal = abs(A[i][i])
        corners = sum(abs(A[i][j]) for j in range(n) if j != i)
        
        if diagonal < corners:
            return False
            
    return True
```

```python

n = int(input("Enter the size of the square matrix (n): "))

print("Please enter the matrix row by row (separate numbers with spaces):")

matrix = []
for i in range(n):
    row = list(map(float, input(f"Row {i+1}: ").split()))
    matrix.append(row)

A = np.array(matrix)

# ---- Output ----
result = diagonally_dominant_check(A)

print("\nIs the matrix diagonally dominant?", result)
print(A)
```
