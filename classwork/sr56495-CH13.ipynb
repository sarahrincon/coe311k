```python
import numpy as np
import time
import matplotlib.pyplot as plt


#naive matrix multiplication
def naive_matrix_multiplication(A,B):
    n = len(A)
    C = [[0] * n for _ in range(n)]
    
    # three nested loops (one for each dimension)
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C
```

```python
sizes = [3, 6, 12, 24, 48, 96, 192, 384]

time_naive = []
time_numpy = []

for n in sizes:
    A = np.random.rand(n, n)
    B = np.random.rand(n, n)
    
    #converting numpy array to list for naive multiplication
    A_list = A.tolist()
    B_list = B.tolist()
    
    #time for naive matrix multiplication
    start = time.time()
    naive_matrix_multiplication(A_list, B_list)
    end = time.time()
    
    time_naive.append(end - start)
    
    #time for numpy
    start = time.time()
    np.dot(A,B)
    end = time.time()
    
    time_numpy.append(end - start)
```

```python
#plotting

plt.figure()
plt.plot(range(1,9), time_naive, label="Naive O(n^3)")
plt.plot(range(1,9), time_numpy, label="Numpy")

plt.xlabel("Matrix Number (1 to 8)")
plt.ylabel("Time to run (in seconds)")
plt.title("Naive vs Numpy Matrix Multiplication")
plt.legend()
plt.grid()

plt.show()
print(f"naive runtime: {time_naive} ")
print(f"numpy runtime: {time_numpy}")
```

As the graph demonstrates, runtime for the naive matrix multiplication grows significantly as the matrix becomes bigger, while the NumPy matrix multiplication shows no significant increase. This might be due to the fact that the naive function uses three nested loops, and results in cubic time complexity. On the other hand, the NumPy function is optimized and does not used nested loops. This means that in this case, for larger matrices, the NumPy is a more efficient tool, since the Naive matrix multiplication grows slower as matrix size increases. However, for smaller matrices, both are fast and efficient.
