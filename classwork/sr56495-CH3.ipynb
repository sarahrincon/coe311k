```python
import random
import math
import matplotlib.pyplot as plt
```

```python
def approximatePi(numSamples):
    #numSamples is the number of random points generated
    counter = 0
    for n in range(numSamples):
        #this generates random points
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
    #this checks if these points fall inside unit circle
        if x*x+y*y<1:
            counter += 1
    #estimates pi by ratio of points inside circle
    return 4 * counter / numSamples
```

```python
#sets number of runs and samples
runs = 100
samples = 1000

absErrors = []
relativeErrors = []

#runs montecarlo 
for i in range(runs):
    estimatePi = approximatePi(samples)
    #calculates absolute error by subtracting pythons pi from our estimate
    absError = abs(estimatePi - math.pi)
    
    relativeError = absError / math.pi
    
    absErrors.append(absError)
    relativeErrors.append(relativeError)
```

```python
plt.plot(absErrors, 'r-o', label='absolute error')
plt.plot(relativeErrors, 'b-o', label='relative error')
plt.xlabel("# of run")
plt.ylabel("error")
plt.title("montecarlo pi error for 100 runs w 1000 samples each")
plt.legend()
plt.grid(True)
plt.show()
```
