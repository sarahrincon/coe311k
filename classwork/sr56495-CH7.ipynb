```python
import numpy as np

#jacobi method
def jacobi(A, b, x0, tol=1e-6, max_iterations=1000):
    n = len(b)
    x = x0.copy()
    x_new = np.zeros_like(x)
    
    for k in range(1, max_iterations+1):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new.copy()
    return x, max_iterations
```

```python
#gauss seidel 

def gauss_seidel(A, b, x0, tol=1e-6, max_iterations=1000):
    n = len(b)
    x = x0.copy()

    for k in range(1, max_iterations+1):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
A1 = np.array([
    [0.582745, 0.48 , 0.10 , 0. , 0. ],
    [0.48 , 1.044129, 0.46 , 0.10 , 0. ],
    [0.10 , 0.46 , 1.10431 , 0.44 , 0.10 ],
    [0. , 0.10 , 0.44 , 0.963889, 0.42 ],
    [0. , 0. , 0.10 , 0.42 , 0.522565]
], dtype=float)

b1 = np.array([1.162745, 2.084129, 2.20431, 1.923889, 1.042565], dtype=float)

x0_1 = np.zeros(len(b1))

x_jacobi_1, it_j1 = jacobi(A1, b1, x0_1)
x_gs_1, it_gs1 = gauss_seidel(A1, b1, x0_1)

print("SYSTEM 1")
print("jacobi:", x_jacobi_1, "iterations:", it_j1)
print("gauss-seidel:", x_gs_1, "iterations:", it_gs1)
print()


A2 = np.array([[1, 1],
               [1, 1.0001]], dtype=float)

b2 = np.array([1, 1], dtype=float)

x0_2 = np.zeros(len(b2))

x_jacobi_2, it_j2 = jacobi(A2, b2, x0_2)
x_gs_2, it_gs2 = gauss_seidel(A2, b2, x0_2)

print("SYSTEM 2")
print("jacobi:", x_jacobi_2, "iterations:", it_j2)
print("gauss-seidel:", x_gs_2, "iterations:", it_gs2)
print()


b3 = np.array([1, 1.0001], dtype=float)

x_jacobi_3, it_j3 = jacobi(A2, b3, x0_2)
x_gs_3, it_gs3 = gauss_seidel(A2, b3, x0_2)

print("SYSTEM 3")
print("jacobi:", x_jacobi_3, "iterations:", it_j3)
print("gauss-seidel:", x_gs_3, "iterations:", it_gs3)
```
