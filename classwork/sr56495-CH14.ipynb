```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: np.sin(x)

#domain
a = 0
b = 2 * np.pi
N = 50

x = np.linspace(a, b, N)
y = f(x)

# step size
h = (b - a) / (N - 1)

#central difference approximation for the derivative
central = (f(x + h) - f(x - h)) / (2 * h)

#forward difference approximation
forward = (f(x + h) - f(x))/ (h)

#backward difference approximation
backward = (f(x) - f(x - h))/ (h)

#plot the approximation and the true derivative
plt.figure()
plt.plot(x, np.cos(x), label = 'exact derivative', linewidth=3)
plt.plot(x, central,
         label='central difference approximation')
plt.plot(x, forward,
         label='forward difference approximation')
plt.plot(x, backward,
         label='backward difference approximation')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("comparison of numerical derivatives")
plt.grid(True)
plt.legend()
plt.show()
```

In the plot above it van be observed how the forward difference approximation best fits the exact derivative and is the most accurate. This makes sense since the forward and backward approximations have more error due to using one-sided information.
