```python
import numpy as np

#data points
x = np.array([1, 2, 3, 4, 5])
y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

#stack the design matrix with a column of ones (for the intercept)
A = np.vstack([x, np.ones(len(x))]).T

#solving using least squares
m, c = np.linalg.lstsq(A, y, rcond=None)[0]
print(f"Slope: {m}, intercept: {c}")
```

```python
#part one

import matplotlib.pyplot as plt

best_fit = m * x + c

plt.figure()
plt.scatter(x, y, label="data points")
plt.plot(x, best_fit, label="best fit line")
plt.xlabel("x")
plt.ylabel("y")
plt.title("least squares best fit (pt1)")
plt.legend()
plt.show()
```

```python
#part two

random_order = np.random.permutation(len(x))
x_randomized = x[random_order]
y_randomized = y[random_order]

A_randomized = np.vstack([x_randomized, np.ones(len(x_randomized))]).T
m_randomized, c_randomized = np.linalg.lstsq(A_randomized, y_randomized, rcond=None)[0]

print(f"randomized slope: {m_randomized}")
print(f"randomized intercept: {c_randomized}")

randomized_best_fit = m_randomized * x_randomized + c_randomized

plt.figure()
plt.scatter(x_randomized, y_randomized, label="data points")
plt.plot(x_randomized, randomized_best_fit, label="best fit line")
plt.xlabel("x")
plt.ylabel("y")
plt.title("least squares after randomizing data (pt2)")
plt.legend()
plt.show()
```

### After randomizing the vectors, the best fit line stays the same

```python
# part 3

x_3 = np.arange(0, 1.01, 0.01)

original_m = 3.5
original_c = 7

noise = np.random.normal(0, 0.1, len(x_3))

y_3 = original_m * x_3 + original_c + noise

#least squares
A_3 = np.vstack([x_3, np.ones(len(x_3))]).T
m_3, c_3 = np.linalg.lstsq(A_3, y_3, rcond=None)[0]

print(f"original slope: {original_m}, intercept:{original_c}")
print(f"estimated slope: {m_3}, intercept: {c_3}")

best_fit_3 = m_3 * x_3 + c_3

plt.figure()
plt.scatter(x_3, y_3, s=10, label="data points")
plt.plot(x_3, best_fit_3, label="least squares fit")
plt.xlabel("x")
plt.ylabel("y")
plt.title("least squares w data pt 3")
plt.legend()
plt.show()
```
