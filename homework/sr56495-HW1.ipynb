```python
import random
import math
import matplotlib.pyplot as plt
```

```python
def approximatePi(numSamples):
    #numSamples is the number of random points generated
    counter = 0
    for n in range(numSamples):
        #this generates random points
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
    #this checks if these points fall inside unit circle
        if x*x+y*y<1:
            counter += 1
    #estimates pi by ratio of points inside circle
    return 4 * counter / numSamples
```

```python
#starts at 10, stops at 1001, jumps in 20
samples = range(10, 1001, 20)
errors = []
for n in samples:
    estimatePi = approximatePi(n)
    #calculates absolute error by subtracting pythons pi from our estimate
    error = abs(estimatePi - math.pi)
    errors.append(error)
    
```

```python
plt.plot(samples, errors, 'r-o')
plt.xlabel("number of samples:")
plt.ylabel("abs error:")
plt.title("montecarlo pi approximation error:")
plt.grid(True)
plt.show()
```
